```python
#csv file handling
import csv
with open('names.csv','r') as a, open('new_names.csv','w') as b:
    csv_reader = csv.reader(a)
    csv_writer = csv.writer(b, delimiter='-')
    
    for line in csv_reader:
        print(line)
        csv_writer.writerow(line)  
```
