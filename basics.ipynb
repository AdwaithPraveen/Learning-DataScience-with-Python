```python
#operators in python

#arithmetic
a = 10
b = 2
print(" + sum",a+b)
print(" - difference:",a-b)
print(" / quotient :",a/b)
print(" '%' remainder :",a%b)

#relational
print("a>b",a>b)
print("a==b",a==b)
print("a<=b",a<=b)
print("a!=b",a!=b)

#logical
print("a and b :",True and False)
print("a or b: ", True or True)
c = True
print("c not: ",not c )

#bitwise 
print("a & b", a & b)
print("a | b ", a | b)
print("a ^ b", a ^ b)
print("~a",~a)
print("a >> 2", a>> 2)
print("a << 2", a<<2)


#assignment
e = 10
f = e
print(" f= ",e)
f = 2
e+=f
print("e+=f :", e)
e-=f
print("e-=f :", e)
e*=f
print("e*=f :", e)
e%=f
print("e%=f :", e)


```

```python
#switch of python - match
num = 10
match num:
    case 1:
        print("fail")
    case 10:
        print("success")
        
name = 'Adwaith' 
match name:
    case 'abc':
        print("wrong username")
    case 'Adwaith':
        print("welcome user")
        
```

```python
#### Functions ####

#default parameter
def abc(x,y=20):
    print(x)
    print(y)
    
abc(20)

#keyword and non keyword args
def xyz(*args, **kwargs):
    for i in args:
        print("args without key :",i)
    for key, value in kwargs.items():
        print(f"args with keys : {key} and {value}")
        
xyz('adwaith', 'abc', name = 'deb', age= 23)

#lambda(func with no name)
        
```
