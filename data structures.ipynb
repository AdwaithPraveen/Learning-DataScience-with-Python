```python
    # Strings
# '' or "" for one line 
# """ """ for multiple lines

s = "WelcomeToPython"
print("1:4",s[1:4])
print("3:",s[3:])
print(":3",s[:3])
print("::-1 reversing", s[::-1])

#editing characters(capitalize first char)
s = "adwaith"
print(s.capitalize())

result = s[0].upper() + s[1:]
print(result)

f = "            Hello world "
Result = f.strip()
print(Result)

#checking if a word is present in or not
print("Hello" in Result)




```

LISTS

```python
#types
a = [1,2,3,4]
b = ['abc','chg']
c = [True, False, 1 , 'adwaith']
print(a,b,c)


#list() function
d = list((1,2,3,'adwaith'))
e = list("Adwaith")
print(d,e)
 
```

```python
#list comprehension
squares = [x**2 for x in range(1,6)]
print(squares)
#for matrix
matrix = [[j*3+i+1 for i in range(3)] for j in range(3)]
for row in matrix:
    print(row)
```
